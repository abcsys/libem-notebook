```python
def plot(precision, recall, f1, latency, cost, name):
    
    plt.rc('font', size=20)          # controls default text sizes
    plt.rc('axes', labelsize=25)     # fontsize of the x and y labels
    plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
    
    # Accuracy
    x = list(f1.keys())
    y1 = list(precision.values())
    y2 = list(recall.values())
    y3 = list(f1.values())

    x_axis = np.arange(len(x)) 
    plt.bar(x_axis - 0.2, y1, 0.2, label = 'Precision', color = '#3CA69B') 
    plt.bar(x_axis, y2, 0.2, label = 'Recall', color = '#F39484') 
    plt.bar(x_axis + 0.2, y3, 0.2, label = 'F1', color = '#FFC906') 
    plt.xticks(x_axis, x) 
    
    plt.legend(ncol=3, fontsize=16, loc='upper center')
    plt.ylim(0, 120)
    plt.yticks((0, 25, 50, 75, 100))

    plt.xlabel("Models")
    plt.ylabel("Accuracy")
    plt.title(name)
    plt.show()
    
    # Latency
    x = list(latency.keys())
    y = list(latency.values())

    bar = plt.bar(x, y)
    bar[0].set_color('#3CA69B')
    bar[1].set_color('#F39484')
    bar[2].set_color('#FFC906')

    plt.xlabel("Models")
    plt.ylabel("Latency")
    plt.title(name)
    plt.show()

    # Cost
    x = list(cost.keys())
    y = list(cost.values())

    bar = plt.bar(x, y)
    bar[0].set_color('#3CA69B')
    bar[1].set_color('#F39484')
    bar[2].set_color('#FFC906')

    plt.xlabel("Models")
    plt.ylabel("Cost")
    plt.title(name)
    plt.show()

for dataset in [amazon_google, dblp_acm, itunes_amazon]:
    plot(dataset[0], dataset[1], dataset[2], dataset[3], dataset[4], dataset[5])
```
